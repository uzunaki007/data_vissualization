```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
#plot a scatter plot.
#import numpy as np
#np.random.seed(3)
#x = 3 + np.random.normal(0, 2, 50)
#y = 3 + np.random.normal(0, 2, len(x))
#ans2.
#(I) scatterplot is a plot which can be created by importing the pyplot module of matplotlib and this figure which
# is used to visualise data, only plots the cartesian coordinates in the graph.
import matplotlib.pyplot as plt
#(II)
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#(III)
plt.scatter(x,y,c='r',alpha=0.7)
plt.xlabel("this is my x-axis")
plt.ylabel("this is my y-axis")
plt.title("this my x vs y graph")
plt.grid()
```

```python
#individual list box1
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box1)
plt.show()
```

```python
#individual list box2
import matplotlib.pyplot as plt
import numpy as np
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
plt.show()
```

```python
#both thelist box1 & box 2
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box3 = [box1,box2]
plt.boxplot(box3,labels=["box1","box2"],patch_artist=True)
plt.show()
```
